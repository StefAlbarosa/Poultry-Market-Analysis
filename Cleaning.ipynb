# Data Merging and Editing

```python
#First I merge population with poultry and then with meat
poul = pd.merge(poul, pop[['Area','Area Code (M49)', 'Year', 'Value']], on=['Area','Area Code (M49)', 'Year'], how='inner')
meat = pd.merge(meat, pop[['Area','Area Code (M49)', 'Year', 'Value']], on=['Area','Area Code (M49)', 'Year'], how='inner')
```

```python
display(poul.columns)
display(meat.columns)
```

```python
poul = poul.rename(columns={'Area': 'Country'})
meat = meat.rename(columns={'Area': 'Country'})
```

## FAO Cleaning

```python
#Editing FAO data
# China regions 
china_regions = ['China, Hong Kong SAR', 'China, Macao SAR', 'China, mainland',
       'China, Taiwan Province of']

# Dropped China's provinces
poul = poul[~poul['Country'].isin(china_regions)]
meat = meat[~meat['Country'].isin(china_regions)]

#Re-naming for merging
poul['Country'] = poul['Country'].replace('Bolivia (Plurinational State of)','Bolivia')
meat['Country'] = meat['Country'].replace('Bolivia (Plurinational State of)','Bolivia')

poul['Country'] = poul['Country'].replace('Iran (Islamic Republic of)','Iran')
meat['Country'] = meat['Country'].replace('Iran (Islamic Republic of)','Iran')

poul['Country'] = poul['Country'].replace('Netherlands (Kingdom of the)','Netherlands')
meat['Country'] = meat['Country'].replace('Netherlands (Kingdom of the)','Netherlands')

poul['Country'] = poul['Country'].replace('Venezuela (Bolivarian Republic of)','Venezuela')
meat['Country'] = meat['Country'].replace('Venezuela (Bolivarian Republic of)','Venezuela')
```

```python
poul.head(2)
```

## KFC Cleaning

```python
#Editing KFC data 
non_countries = ['Bonaire', 'Curacao', 'Gibraltar', 'Grand Cayman', 'Military - Africa',
                 'Military - Asia', 'Military - Europe', 'Military - Latin America',
                 'Military - MENA', 'Saipan', 'StKitts', 'StLucia', 'StMaarten',
                 'StVincent', 'USVI']

# Drop rows corresponding to non-country entries from kfc2016
kfc2016 = kfc2016[~kfc2016['Country'].isin(non_countries)]
kfc2018 = kfc2018[~kfc2018['Country'].isin(non_countries)]

# Re-naming 
kfc2016['Country'] = kfc2016['Country'].replace('Antigua', 'Antigua and Barbuda')
kfc2016['Country'] = kfc2018['Country'].replace('Antigua', 'Antigua and Barbuda')
```

```python
#only keeping columns needed. Not keeping Country because later we will need all columns as numbers for the PCA
poul=poul[['Area Code (M49)', 'Country','Element', 'Item', 'Year', 'Value_x', 'Value_y']]
meat=meat[['Area Code (M49)','Country', 'Element', 'Item', 'Year', 'Value_x', 'Value_y']]
```

## FAO Pivot Items

```python
#creating the pivot tables:
    
poul = pd.pivot_table(poul, 
                    index=['Area Code (M49)','Country', 'Item', 'Year', 'Value_y'], 
                    columns=['Element'], 
                    values=['Value_x']).reset_index()

meat = pd.pivot_table(meat, 
                    index=['Area Code (M49)','Country', 'Item', 'Year', 'Value_y'], 
                    columns=['Element'], 
                    values=['Value_x']).reset_index()
```

```python
display(poul.columns)
display(meat.columns)
```

```python
#poultry data:
# I am resetting the column names and making it into single dimension.
poul.columns = poul.columns.droplevel(1)
poul.set_index(['Area Code (M49)','Country'], inplace=True)
poul = poul.rename_axis(index={'Area Code (M49)': 'area_c','Country': 'country'})

# population, domestic supply and import q are in 1000s 
poul.columns = ['item', 'year', 'population','p_domestic_supply','p_food_supply_(kg/cap/day)', 'p_import', 'p_protein_supply_(g/cap/day)']


#meat data:
# I am resetting the column names and making it into single dimension.
meat.columns = meat.columns.droplevel(1)
meat.set_index(['Area Code (M49)','Country'], inplace=True)
meat = meat.rename_axis(index={'Area Code (M49)': 'area_c','Country': 'country'})

# population, domestic supply and import q are in 1000s 
meat.columns = ['item', 'year', 'population','m_domestic_supply','m_food_supply_(kg/cap/day)', 'm_import', 'm_protein_supply_(g/cap/day)']
```

```python
display(poul.head(2))
display(meat.head(2))
```

```python
poul.loc[:, 'p_import_(kg/cap/day)'] = (poul['p_import'] * 1000000)/ (poul['population']*1000)
meat.loc[:, 'm_import_(kg/cap/day)'] = (meat['m_import']* 1000000) / (meat['population'] *1000)

poul.loc[:, 'p_domestic_supply_(kg/cap/day)'] = (poul['p_domestic_supply'] * 1000000)/ (poul['population']*1000)
meat.loc[:, 'm_domestic_supply_(kg/cap/day)'] = (meat['m_domestic_supply']* 1000000) / (meat['population'] *1000)
```

```python
#only keeping essential columns for later on making the annnual calculations
poul_s = poul[['year', 'population', 'p_import_(kg/cap/day)', 'p_domestic_supply_(kg/cap/day)','p_protein_supply_(g/cap/day)']]
meat_s = meat[['year', 'population', 'm_import_(kg/cap/day)', 'm_domestic_supply_(kg/cap/day)','m_protein_supply_(g/cap/day)']]
```

## % Calculations

```python
display(poul_df.columns)
display(meat_df.columns)
```

```python
p_elements = 'p_domestic_supply_(kg/cap/day)', 'p_protein_supply_(g/cap/day)','p_import_(kg/cap/day)'
m_elements ='m_domestic_supply_(kg/cap/day)', 'm_protein_supply_(g/cap/day)','m_import_(kg/cap/day)'

for p in p_elements:
    poul_df[(p, f'ds%_2016_2018')] = (poul_df[(p, '2018')] - poul_df[(p, '2016')]) / poul_df[(p, '2016')] * 100

for m in m_elements:
    meat_df[(m, f'ds%_2016_2018')] = (meat_df[(m, '2018')] - meat_df[(m, '2016')]) / meat_df[(m, '2016')] * 100
```

```python
# Calculating % change for KFC only on poultry because I will merge the poultry and meat later
df[('kfc', 'kfc%_2016_2018')] = ((df[('kfc', '2018')] - df[('kfc', '2016')]) / df[('kfc', '2016')]) * 100
```

## Null, NaN, inf values

```python
# First, I replaced the NaN values for 0 in the kfc store columns, because where there is not data for it, 
# we can assume its 0. 
df[('kfc', '2016')].fillna(0, inplace=True)
df[('kfc', '2018')].fillna(0, inplace=True)
```

```python
#Checking for NaNs 
display(df[df.isna().any(axis=1)].head())

#Seems like we get NaNs in the the % columns if both 2016 and 2018 are 0 for the given item.
```

```python
# Checking for inf
df[np.isinf(df).any(axis=1)].head(5)
#Seems like we get inf we the division value (so 2016) is 0
```

```python
# I want to replace NaN values in the kfc % and import% columns because they appear when both 2016 and 2018 columns are 0
df[('kfc', 'kfc%_2016_2018')].fillna(0, inplace=True)
df[('p_import_(kg/cap/day)', 'ds%_2016_2018')].fillna(0, inplace=True)
df[('m_import_(kg/cap/day)', 'ds%_2016_2018')].fillna(0, inplace=True)
```

```python
# No more NaN values
df[df.isna().any(axis=1)]
```

```python
# Function 1 for kfc% inf values.
def kfc_inf_values(df):
    for index, row in df.iterrows():
        if np.isinf(row[('kfc', 'kfc%_2016_2018')]):
            df.at[index, ('kfc', 'kfc%_2016_2018')] = row[('kfc', '2018')] * 100
    return df

# Function 2 for poultry import % inf values
def p_import_inf_values(df):
    for index, row in df.iterrows():
        if np.isinf(row[('p_import_(kg/cap/day)',  'ds%_2016_2018')]):
            df.at[index, ('p_import_(kg/cap/day)',  'ds%_2016_2018')] = row[('p_import_(kg/cap/day)','2018')] * 100
    return df

# Function 2 for meat import % inf values
def m_import_inf_values(df):
    for index, row in df.iterrows():
        if np.isinf(row[('m_import_(kg/cap/day)',  'ds%_2016_2018')]):
            df.at[index, ('m_import_(kg/cap/day)',  'ds%_2016_2018')] = row[('m_import_(kg/cap/day)','2018')] * 100
    return df

# Applying Functions:
functions = kfc_inf_values, p_import_inf_values, m_import_inf_values

for function in functions:
    df = function(df)
```
